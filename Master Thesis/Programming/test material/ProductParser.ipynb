open and work with xml via ElementTree

```python
import xml.etree.ElementTree as ET
root = ET.parse('products.xml').getroot()
print(root)
```

convert xml file into in csv

```python
csv = "NumberProducts;Buyer;Price;"
for row in root.findall('table/row'):
    for cell in row.findall('cell'):
        csv = csv + str(cell.text) + ";"
    csv = csv + "\n"
    
print(csv)
```

save CSV file

```python
f = open("products.csv", "w", encoding="utf-8")
f.write(csv)
f.close()
```

Takes Pandas and reads csv to set Number as Index for CSV

```python
import pandas as pd
df = pd.read_csv('products.csv', sep=';', header=0)
print(df)
```

Uses Pandas to convert float values to integer

```python
data = pd.DataFrame.to_csv(self=df, sep=";", float_format='%.f', mode='w', encoding='utf-8')
print(data)
```

saves changes in new file

```python
f = open('productsN.txt', 'w', encoding='utf-8')
f.write(data)
f.close()
```

start to extract products and prices for horses and save them in a new csv & txt file

```python
import re
#input('Please write the product you\'re searching for:')
f = open('productsN.txt', 'r', encoding='mbcs')
word = ['(.*paard.*)|(.*veul.*)|(.*merrij.*)|(.*hengste.*)']
for line in f:
    for i in word:
        if re.search(i, line):
            print(line)
            horses = line
            horsesFile = open('horses.csv', 'a', encoding='utf-8')
            horsesFile.write(horses)
            horsesFile.close() 
            horsesFileTxt = open('horses.txt', 'a', encoding='utf-8')
            horsesFileTxt.write(horses)
            horsesFileTxt.close()
```
